```python
def plot_map(var, year, commodity=None):
    data_ = data[(data['Variable'] == var) & (data['Year'] == year)]
    if commodity != None:
        data_ = data_[(data_['Commodity'] == commodity)]

    merged = world.merge(data_, left_on='NAME', right_on='Country')

    # Define bins and labels
    m = merged['Value'].max()
    bins = [0, 0.2*m, 0.4*m, 0.6*m, 0.8*m, m]
    labels = [f"< {round(0.2*m, 1)}", f"{round(0.2*m, 1)} - {round(0.4*m, 1)}", 
              f"{round(0.4*m, 1)} - {round(0.6*m, 1)}", f"{round(0.6*m, 1)} - {round(0.8*m, 1)}", 
              f"> {round(0.8*m, 1)}"]
    merged['ValueClass'] = pd.cut(merged['Value'], bins=bins, labels=labels)


    cmap = ListedColormap(['#59A1E3', '#A8CFEA', '#D4C6BA', '#D5B08D', '#DC9554'])
    
    fig, ax = plt.subplots(1, 1, figsize=(8, 8))
    ax.set_facecolor('#EBECEC')

    merged.plot(column='ValueClass', 
                ax=ax,
                edgecolor='grey',
                linewidth=0.8, 
                legend=True,
                cmap=cmap)
    ax.set_title(f'{var} in {year}')

    for x, y, label in zip(merged['LON'], merged['LAT'], merged['Value']):
        ax.text(x, y, round(label, 2), fontsize=6, c='black', 
                ha='center', va='center')

    plt.show()  
```

```python
# - economic data per multiple countries over time (line plot)
def ec_lines(var, countries):
        
        data_ = data[(data['Variable'] == var) & (data['Country'].isin(countries))]
        data_ = data_.pivot_table(index='Year', columns='Country', values='Value', aggfunc='mean')
        
        ax = data_.plot(colormap='Accent', legend=False, figsize=(10, 5))
        ax.set_facecolor('#EBECEC')
        ax.grid(True) 
        
        for line, name in zip(ax.get_lines(), data_.columns):
                y = line.get_ydata()[0]
                ax.annotate(name, xy=(0, y+7), xytext=(5, 0), 
                        xycoords=('axes fraction', 'data'), textcoords='offset points',
                        color="black", fontsize=9, va='center')
        
        plt.title(f'{var} of {countries}')
        plt.show()
```

```python
# - one metric of multiple commodities per single country over time  (line plot)
def con_lines(var, country, commodities = None):
        
    data_ = data[(data['Variable'] == var) & (data['Country'] == country)]
    if commodities != None:
        data_ = data_[(data_['Commodity'].isin(commodities))]
    data_ = data_.pivot_table(index='Year', columns='Commodity', values='Value')
        
    ax = data_.plot(colormap='Accent', legend=False, figsize=(10, 5))
    ax.set_facecolor('#EBECEC')
    ax.grid(True) 
    
    for line, name in zip(ax.get_lines(), data_.columns):
            y = line.get_ydata()[0]
            ax.annotate(name, xy=(0, y+0), xytext=(2, 0), 
                    xycoords=('axes fraction', 'data'), textcoords='offset points',
                    color="black", fontsize=10, va='center')
    
    plt.title(f'{var} in {country}')
    plt.show()
```

```python
def scatter_plot(countries, years, var1, var2, com1=None, com2=None):
    data_ = data[(data['Country'].isin(countries)) & (data['Year'].isin(years))]
    data_ = data_[(data_['Variable'].isin([var1, var2]))]
    if com1 != None:
        index1 = data_[(data_['Variable'] == var1) & (data_['Commodity'].isin(com1))].index
    else:
        index1 = data_[(data_['Variable'] == var1)].index
    
    if com2 != None:
        index2 = data_[(data_['Variable'] == var2) & (data_['Commodity'].isin(com2))].index
    else:
        index2 = data_[(data_['Variable'] == var2)].index
    
    data_ = data_.loc[index1.union(index2)]
    data_ = data_.pivot_table(index=['Region', 'Country', 'Year'], columns='Variable', values='Value', aggfunc='mean').reset_index()

    regions = data_['Region'].unique()
    cmap = plt.get_cmap('tab20', len(regions))
    region_colors = {region: cmap(i) for i, region in enumerate(regions)}

    fig, ax = plt.subplots(figsize=(10, 5))
    ax.set_facecolor('#EBECEC')
    ax.grid(True)

    for region in regions:
        subset = data_[data_['Region'] == region]
        ax.scatter(subset[var1], subset[var2], label=region, color=region_colors[region])
    
    plt.title(f'Yearly {var1} and {var2} of apples in Europe')
    plt.xlabel(var1)
    plt.ylabel(var2)
    plt.legend(title='Region')
    plt.show()
```

## Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data.head()
```

We want to build a model which is able to predict the consumption of a certain commodity based on the economic data for GDP, FCE and salary earners. We will use a multiple linear regression for each commodity and a wrapper function for model usage.

First, we will have to bring the data into the right format:

```python
model_data = data[data['Variable'].isin(['consumption', 'GDP', 'FCE', 'salary earners'])]
model_data['Variable'] = np.where(model_data['Variable'] == 'consumption', model_data['Variable'] + "/" + model_data['Commodity'], model_data['Variable'])
model_data = model_data.pivot_table(index=['Country', 'Year'], columns='Variable', values='Value').reset_index()
```

```python
def train_test_split(data, commodity):
    data_ = data[['Country', 'Year', 'GDP', 'FCE', 'salary earners', f'consumption/{commodity}']]
    data_ = data_.dropna()
    print(data_)
    
    latest_year = data_['Year'].max()
    train = data_[data_['Year'] < latest_year]
    test = data_[data_['Year'] == latest_year]

    X_train = train[['GDP', 'FCE', 'salary earners']]
    y_train = train[f'consumption/{commodity}']
    X_test = test[['GDP', 'FCE', 'salary earners']]
    y_test = test[f'consumption/{commodity}']   
    
    return X_train, y_train, X_test, y_test
```

```python
class ConsumptionModel:
    def __init__(self, data, commodity):
        self.commodity = commodity
        self.model = LinearRegression()
        self.X_train, self.y_train, self.X_test, self.y_test = train_test_split(data, commodity)
        self.y_train_p = None
        self.y_test_p = None
        self.rmse = None
        self.rmse_rel = None
        self.r2 = None
        self.coeff_impact = None
    
    def fit(self):
        self.model.fit(self.X_train, self.y_train)
        sum = np.sum(np.abs(self.model.coef_))
        self.coeff_impact = self.model.coef_ / sum
        
        print('Intercept:', self.model.intercept_)
        print('Coefficients:', self.model.coef_)
    
    def predict(self):
        self.y_train_p = self.model.predict(self.X_train)
        self.y_test_p = self.model.predict(self.X_test)
    
    def metrics(self):
        if self.y_test_p is None:
            self.predict()
        mean = self.y_test.mean()
        self.rmse = np.sqrt(mean_squared_error(self.y_test, self.y_test_p))
        self.rmse_rel = self.rmse / mean
        self.r2 = self.model.score(self.X_test, self.y_test)
        
        print(f'RMSE: {self.rmse}')
        print(f'RMSE relative to mean: {self.rmse_rel}')
        print(f'R2: {self.r2}')
    
    def plot(self):
        plt.figure(figsize=(10, 6))
        plt.scatter(self.y_train, self.y_train_p, label='Training Data', alpha=0.6, color='#606c38')
        plt.scatter(self.y_test, self.y_test_p, label='Validation Data', alpha=0.8, color='#bc6c25')
        plt.plot([self.y_train.min(), self.y_train.max()], [self.y_train.min(), self.y_train.max()], color='#283618', linestyle='--')
        plt.title(f'{self.commodity}-model: actual vs predicted values')
        plt.xlabel('Actual')
        plt.ylabel('Predicted')
        plt.legend()
        plt.gca().set_facecolor('#fefae0')
        plt.savefig(f'assets/{self.commodity}_model.png')
        plt.show()
        
    def plot_coefficients(self):
        palette = ['#606c38', '#bc6c25', '#283618']
        plt.figure(figsize=(10, 3))
        plt.barh(self.X_train.columns, self.coeff_impact, color=palette)
        plt.title(f'{self.commodity}-model: coefficients')
        plt.gca().set_facecolor('#fefae0')
        plt.xlim(left=min(self.coeff_impact) - 0.1, right=max(self.coeff_impact) + 0.1)
        for i, coef in enumerate(self.coeff_impact):
            plt.text(coef, i, round(coef, 2), ha='left' if coef > 0 else 'right', va='center')
        plt.savefig(f'assets/{self.commodity}_coefficients.png')
        plt.show()
```

```python
def pipeline(commodity, plot=True, return_model=False):
    model = ConsumptionModel(model_data, commodity)
    model.fit()
    model.predict()
    model.metrics()
    if plot:
        model.plot()
        model.plot_coefficients()
    if return_model:
        return model
```

```python
pipeline('apples')
```

```python
# list all commodities except 'nan'
commodities = data['Commodity'].unique()
commodities = commodities[~pd.isnull(commodities)]
commodities
```

```python
results = pd.DataFrame(columns=['Commodity', 'RMSE', 'RMSE_rel', 'R2', 'Intercept', 'GDP_beta_rel', 'FCE_beta_rel', 'SalaryEarners_beta_rel'])
for commodity in commodities:
    print(f"Linear Regression Model for {commodity}")
    model = pipeline(commodity, plot=False, return_model=True)
    row = {'Commodity': commodity, 
           'RMSE': model.rmse,
           'RMSE_rel': model.rmse_rel,
           'R2': model.r2, 
           'Intercept': model.model.intercept_,
           'GDP_beta_rel': model.coeff_impact[0], 
           'FCE_beta_rel': model.coeff_impact[1], 
           'SalaryEarners_beta_rel': model.coeff_impact[2]}
    results.loc[len(results)] = row
    print('')
    
results
```

```python
mean_coefficients = results[['GDP_beta_rel', 'FCE_beta_rel', 'SalaryEarners_beta_rel']].mean()
palette = ['#606c38', '#bc6c25', '#283618']
plt.figure(figsize=(10, 3))
plt.barh(mean_coefficients.index, mean_coefficients.values, color=palette)
plt.title('Mean Coefficients of GDP, FCE, and Salary Earners')
plt.gca().set_facecolor('#fefae0')
plt.xlim(left=min(mean_coefficients.values) - 0.01, right=max(mean_coefficients.values) + 0.01)
for i, coef in enumerate(mean_coefficients.values):
    plt.text(coef, i, round(coef, 3), ha='left' if coef > 0 else 'right', va='center')
plt.show()
```
